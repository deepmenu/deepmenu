```python
import numpy as np
import pandas as pd
import re
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
oe = pd.read_csv('Data/OrderEntry.txt')
```

```python
oe[oe['Restaurant']==10]['Food'].mean()
```

```python
oem = oe.groupby('Restaurant').mean()['Food']
print(oem.shape)
oem.head()
```

```python
plt.figure(num=None, figsize=(2, 10), dpi=80, facecolor='w', edgecolor='k')
plt.rcParams.update({'font.size': 20})
# plt.ylabel('Average price', fontsize=22)
# x = np.zeros_like(oem)
x = np.random.randn(len(oem))
plt.plot(x, oem, '+', color='r')
plt.xticks([])
plt.title('Average price among restaurants')
plt.savefig('price.png')
plt.show()
```
