```python
import numpy as np
import pandas as pd
import re
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
oe = pd.read_csv('Data/OrderEntry.txt')
oe.head()
```

```python
date_vec = pd.DataFrame(oe['TimeStart'])
date_vec.head()
```

```python
datelist = [s[-8:-6] for s in date_vec['TimeStart']]
# datelist
```

```python
date_vec = pd.DataFrame(datelist, columns=['Hour'])
date_vec = date_vec['Hour'].value_counts()
date_vec = date_vec.sort_index()
# sns.distplot(date_vec)
```

```python
plt.figure(num=None, figsize=(15, 10), dpi=80, facecolor='w', edgecolor='k')
plt.rcParams.update({'font.size': 20})
# plt.figure.suptitle('test title', fontsize=20)
plt.xlabel('Time', fontsize=24)
plt.ylabel('Number of orders', fontsize=24)
plt.title('Most frequent order time')
plt.bar(date_vec.index, date_vec, color='g') #[0].set_color('r')
# plt.gca().invert_yaxis()
plt.savefig('hours.png')
plt.show()
```
