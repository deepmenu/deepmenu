```python
import numpy as np
import pandas as pd
import re
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
oe = pd.read_csv('Data/OrderEntry.txt')
oe.head()
```

```python
date = pd.to_datetime(oe['TransDate'])
date
```

```python
date = date.apply(lambda x: x.strftime('%A')).value_counts()
date
```

```python
plt.figure(num=None, figsize=(15, 10), dpi=80, facecolor='w', edgecolor='k')
plt.rcParams.update({'font.size': 20})
plt.xlabel('Day of week', fontsize=24)
plt.ylabel('Number of orders', fontsize=24)
plt.title('Most frequent order days')
plt.bar(date.index, date, color='g')
# plt.gca().invert_yaxis()
plt.savefig('week.png')
plt.show()
```
